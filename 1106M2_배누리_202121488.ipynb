# 결측데이터 처리하기

## 결측데이터 확인

```python
# 데이터 불러오기
import pandas as pd
filename='/content/sample_data/train.csv'
df=pd.read_csv(filename)
df.head()
```

```python
# 연습 - 결측데이터('Cabin') 확인하기
#자료가 많으므로 .head() 설정

df[df['Cabin'].isna()] .head()
```

## 결측데이터 삭제

```python
#drop 함수 이용
#Cabin 컬럼 전체 삭제

df.drop('Cabin',axis=1,inplace=True)
```

```python
#컬럼 이름 확인
df.columns
```

```python
#삭제된 데이터 프레임 확인

df.head()
```

## 결측데이터 대체 : 최빈값으로 채우기

```python
# 최빈값 찾기
# 'Embarked' 컬럼에 가장 빈번한 값은 S이고, S를 인덱싱 0으로 지정.
df['Embarked'].mode()
```

```python
#Embarked 컬럼의 결측데이터를 인덱싱 0 (값S로 대체)

df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)
df.head()
```

```python
#Embarked 컬럼의 결측데이터 확인
df[df['Embarked'].isna()]
```

```python
# 결측데이터가 대체된 데이터 프레임의 전체 결측데이터 확인
df.isnull().sum()
```

```python
# 결측데이터 대체되기 전 결측데이터 포함 행들 출력 및 확인
df.loc[61, 'Embarked']
```

```python
df.loc[829, 'Embarked']
```

```python
# 최빈값 찾기
# 'Age' 컬럼에 가장 빈번한 값은 S이고, S를 인덱싱 0으로 지정.
df['Age'].mode()
```

```python
#연습 - Age 컬럼의 결측데이터를 인덱싱 0 값S로 대체
#상위 10행을 출력하여, 인덱싱 5의 결측데이터 대체상태 확인

df['Age'].fillna(df['Age'].mode()[0], inplace=True)
df.head(10)
```

```python
#연습 - Age 컬럼의 결측데이터 확인

df[df['Age'].isna()]
```

```python
# 결측데이터가 대체된 데이터 프레임의 전체 결측데이터 확인
df.isnull().sum()
```

```python
# 결측데이터 대체되기 전 결측데이터 포함 행들 출력 및 확인
#5,17,19 인덱싱

df.loc[5, 'Age']
```
